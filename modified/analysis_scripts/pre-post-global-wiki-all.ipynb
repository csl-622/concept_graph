```python
import networkx as nx
import matplotlib.pyplot as plt
import os
import math
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
from collections import Counter
import pickle
import nltk.data
import errno
import time
from nltk.corpus import stopwords
nltk.download('stopwords')

def preProcessing(S):
    ps = PorterStemmer()
    stop_words = set(stopwords.words('english'))
    tokenizer = RegexpTokenizer(r'[a-zA-Z]+')
    line = tokenizer.tokenize(S)
    if line != []:
        L = []
        for j in range(len(line)):
            x = line[j].lower()
            if x not in stop_words:
                L.append(ps.stem(x))
        L = ' '.join(L)
        L += ' '
    return L

Dict = nx.read_gpickle('dict.pickle')

Id2Word = {}

for key, val in Dict.items():
    Id2Word[val] = key


G = nx.read_gpickle('concept_graph2.gpickle')
nodes_set = set(G.nodes())
G = G.to_undirected()
```

```python
Articles = ['Andhadhun','Euclid']

Colors = ['r']
cols = len(Colors)

for article in Articles:
    if os.path.exists(os.path.join(article,'plots')) is False:
        os.system('mkdir ' + article + '/plots/')
        time.sleep(2)
        os.system('mkdir ' + article + '/plots/pre/')
        os.system('mkdir ' + article + '/plots/post/')
        os.system('mkdir ' + article + '/plots/global/')
        os.system('mkdir ' + article + '/plots/log_pre/')
        os.system('mkdir ' + article + '/plots/log_post/')
        os.system('mkdir ' + article + '/plots/log_global/')
    
    Revisions = os.listdir(article+'/Processed/')
    n = len(Revisions)
    counts = n
    for q in range(1,counts+1):
        p = int((q*n)/counts)
        print(q,p)
        S = ''
        with open(article+'/Processed/'+str(p)+'.txt','r') as file:
            for line in file:
                S += line
        S = S.split(' ')

        H = nx.Graph()
        D = dict()
        article_dict = dict()
        for i in S:
            if i not in article_dict:
                if i in Dict:
                    x = Dict[i]
                    if x in nodes_set:
                        article_dict[i] = x
                        H.add_node(x)
                        D[x] = (0,0,G.degree(x))


        N = list(H.nodes())
        for i in range(len(N)):
            for j in range(i+1,len(N)):
                if G.has_edge(N[i],N[j]) is True:
                    H.add_edge(N[i],N[j])


        curr_nodes_set = set()
        new_nodes_set = set(H.nodes())
        I = nx.Graph()
        for i in S:
            if i in article_dict:
                j = article_dict[i]
                if j in new_nodes_set:
                    if j not in curr_nodes_set:
                        curr_nodes_set.add(j)
                        I.add_node(j)
                        for k in H.neighbors(j):
                            if k in curr_nodes_set:
                                I.add_edge(j,k)
                        D[j] = (I.degree(j),H.degree(j),D[j][2])


        X = []
        Y = []
        Z = []
        LX = []
        LY = []
        LZ = []
        T = []
        c = 0
        for i,j in D.items():
            c += 1
            T.append(c)
            X.append(j[0])
            Y.append(j[1])
            Z.append(j[2])

            if(j[0]!=0):
                LX.append(math.log(j[0]))
            else:
                LX.append(-1)
                
            if(j[1]!=0):
                LY.append(math.log(j[1]))
            else:
                LY.append(-1)

            if(j[2]!=0):
                LZ.append(math.log(j[2]))
            else:
                LZ.append(-1)
                
        index = q%cols
        col = Colors[index]


        plt.plot(T,sorted(X),col+'-')
        plt.ylabel('Pre Degree --->')
        plt.xlabel('Node index --->')
        plt.savefig(article +'/plots/pre/' + str(p) + '.png', bbox_inches='tight')

        plt.clf()

        plt.plot(T,sorted(Y),col+'-')
        plt.ylabel('Post Degree --->')
        plt.xlabel('Node index --->')
        plt.savefig(article +'/plots/post/' + str(p) + '.png', bbox_inches='tight')

        plt.clf()

        plt.plot(T,sorted(Z),col+'-')
        plt.ylabel('Global Degree --->')
        plt.xlabel('Node index --->')
        plt.savefig(article +'/plots/global/' + str(p) + '.png', bbox_inches='tight')

        plt.clf()

        plt.plot(T,sorted(LX),col+'-')
        plt.ylabel('Log Of Pre Degree --->')
        plt.xlabel('Node index --->')
        plt.savefig(article +'/plots/log_pre/' + str(p) + '.png', bbox_inches='tight')

        plt.clf()

        plt.plot(T,sorted(LY),col+'-')
        plt.savefig(article +'/plots/log_post/' + str(p) + '.png', bbox_inches='tight')

        plt.clf()

        plt.plot(T,sorted(LZ),col+'-')
        plt.savefig(article +'/plots/log_global/' + str(p) + '.png', bbox_inches='tight')

        plt.clf()
```
